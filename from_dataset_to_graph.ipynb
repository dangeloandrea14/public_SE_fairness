```python
import pandas as pd
import os
import numpy as np
from tqdm import tqdm
import ast
import gender_guesser.detector as gender
```

```python
folder = os.path.join("processed_data")
```

```python
total = pd.read_csv(os.path.join(folder, 'total.csv'), index_col = 0)
```

## Edges (Networkx ready)

```python
total = total.reset_index(drop=True)
```

```python
total
```

```python
total.loc[total.ID == 2120]
```

```python
edges = pd.DataFrame()
coauthors_df = pd.DataFrame()
for idx,row in tqdm( total.iterrows(), total = total.shape[0] ):
    person = (pd.DataFrame( total.loc[idx] ).T )[['ID','Sex','co-authors']]

    coauthors = total.loc[idx,'co-authors']
    
    coauthors =  list(ast.literal_eval(coauthors)) 

    if len(coauthors) < 1:
        continue 
    for c in coauthors:
        te = total.loc[total.ID == c]
        te = te[['ID', 'Sex']].rename(columns={'ID':'ID_y','Sex':'Sex_y'})
        te['ID_x'] = idx
        te['Sex_x'] = row['Sex']

        coauthors_df = pd.concat([coauthors_df,te])


edges = coauthors_df

edges = edges[edges['ID_x'] != edges['ID_y']]
edges = edges[['ID_x','Sex_x','ID_y','Sex_y']]
```
